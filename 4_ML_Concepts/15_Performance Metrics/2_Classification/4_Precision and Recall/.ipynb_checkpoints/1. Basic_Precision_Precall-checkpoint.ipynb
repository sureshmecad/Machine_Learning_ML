- TPR ---> True positive rate


- FPR ---> False positive rate


- TNR ---> Ture negative rate


- FNR ---> False negative rate

- **TP(True Positive)**

      Actual value is 1
      Predicted values is 1.
      
      
- **FN(False Negative)**

      Actual value is 1
      Predicted values is 0.
      
      
- **FP(False Positive)**

      Actual value is 0
      Predicted values is 1.
      
      
- **TN(True Negative)**

      Actual value is 0
      Predicted values is 0.
