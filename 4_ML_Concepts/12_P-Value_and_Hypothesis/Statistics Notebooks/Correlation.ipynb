# Correlation

$$r = \frac{\sum(x-\bar{x})(y-\bar{y})}{\sqrt{(x-\bar{x})^2(y-\bar{y})^2}}$$

## There can be 5 degree of Correlation

1. Strong Negative Correlation
2. Moderate Negative Correlation
3. Strong Positive Correlation
4. Moderate Positive Correlation
5. No Correlation

```python
import scipy.stats as st
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data=pd.read_excel('Sample - Superstore.xls')
```

# Correlation using Numpy

```python
np.corrcoef(data['Sales'],data['Profit'])
```

## Using Scipy

```python
st.pearsonr(data['Sales'],data['Profit'])
```

```python
sns.scatterplot(data['Sales'],data['Profit'])
```

```python
sns.pairplot(data)
```

```python
data.corr()
```

```python
sns.heatmap(data.corr())
```

```python
plt.figure(figsize=(15,8))
plt.subplot(231)
sns.scatterplot(data['Sales'],data['Profit'])
plt.subplot(232)
sns.scatterplot(data['Discount'],data['Profit'])
plt.subplot(233)
sns.scatterplot(data['Quantity'],data['Profit'])
plt.subplot(234)
sns.scatterplot(data['Discount'],data['Sales'])
plt.subplot(235)
sns.scatterplot(data['Discount'],data['Quantity'])
```

```python
data.corr()
```
