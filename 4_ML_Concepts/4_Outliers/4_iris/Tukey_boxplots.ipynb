### Outlier Analysis

- Detection with **Univariate Methods** (Univariate Analysis)

     **1. Tukey boxplots** 


![4_iris](image/1.JPG)

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv("iris.csv", index_col=0)
df.head()
```

```python
plt.figure(figsize=(8,7))
df.boxplot(return_type='dict')
plt.show()
```

![4_iris](image/2.JPG)

```python
iris_outliers = (df['Sepal.Width']>4)
df[iris_outliers]
```

```python
iris_outliers = (df['Sepal.Width']<2.05)
df[iris_outliers]
```

```python
pd.options.display.float_format = '{:.1f}'.format
x_df = pd.DataFrame(df)
print(x_df.describe())
```
