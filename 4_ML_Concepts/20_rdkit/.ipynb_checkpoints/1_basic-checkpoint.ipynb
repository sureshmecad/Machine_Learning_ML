```python
smiles_list = ['CC12CCC3C(C1CCC2=O)CC=C4C3(CCC(C4)O)C', 'C[N+](C)(C)CC(CC(=O)O)O', 'C(CC(=O)O)CN', '[Cl-]', 'C(=O)(O)O',
               'C1C(N(C2=C(N1)NC(=NC2=O)N)C=O)CNC3=CC=C(C=C3)C(=O)NC(CCC(=O)O)C(=O)O', 'C1C2C(C(S1)CCCCC(=O)O)NC(=O)N2']

mol_list = []
for smiles in smiles_list:
    mol = Chem.MolFromSmiles(smiles)
    mol_list.append(mol)
    
img = Draw.MolsToGridImage(mol_list) #, molsPerRow=3
img
```
