```python
# importing necessary packages
import numpy as np
import pandas as pd
import datetime as dt
```

```python
data = pd.read_csv('Child_data.csv')
data.head()
```

```python
data = pd.read_csv('Child_data.csv', index_col=0)
data.head()   # index_col=0 is used because not use the first column while loading the data
```

```python
# Basic checks
data.shape
```

```python
data.isnull().sum() # NaN check
```

```python
data.info() # Column data type check
```

```python
# Calculate Height in cms
ht_in_cms=[]
for ht in data.Height :
    sData = ht.split("'")    # split feet and inches value
    ht_in_cms.append((int(sData[0]) * 12 + int(sData[1])) * 2.54) # convert feet to inches and then multiply by 2.54

ht_in_cms
```

```python
# Create new height column    
data['Height (in cms)'] = ht_in_cms
data.head()
```

```python
# Determine Gender, Bigger Foot and Dominant Hand
gender=[]
biggerFoot=[]
dominantHand=[]
for r in data.Remarks :
    if 'Girl' in r :                                
        gender.append('Female')
    elif 'Boy' in r :  
        gender.append('Male')
    else :
        gender.append('Nan')
    
    if 'Left biggerfoot' in r :
        biggerFoot.append('L')
    elif 'Right biggerfoot' in r :  
        biggerFoot.append('R')
    else :
        biggerFoot.append('Nan')
        
    if 'Left dominant' in r :
        dominantHand.append('L')
    elif 'Right dominant' in r :  
        dominantHand.append('R')
    else :
        dominantHand.append('Nan')

data['Gender'] = gender
data['Bigger Foot'] = biggerFoot
data['Dominant Hand'] = dominantHand

data.tail()
```

```python
data.isnull().sum() # checking if any NaN was introduced
```

```python
# spliting Day to week number and week day
week=[]
day=[]
for d in data.Day:
    sData = d.split(' ')
    week.append(sData[0])
    day.append(sData[1])
data['Week'] = week
data['Day'] = day

data.head(2)
```

```python
# Converting Day to numerical representation
days = {'Monday':0,'Tuesday':1,'Wednesday':2,'Thursday':3,'Friday':4,'Saturday':5,'Sunday':6}
index =1
for d in data.Day :
    data.Day[index] = days[d]
    index += 1
```

```python
# Converting Week number to numerical representation
data.Week.replace(to_replace='1st',value=1,inplace=True)
data.Week.replace(to_replace='2nd',value=2,inplace=True)
data.Week.replace(to_replace='3rd',value=3,inplace=True)
data.Week.replace(to_replace='4th',value=4,inplace=True)

data.head(2)
```

```python
# changing year from yy to yyyy
data['Birth Year'] = data['Birth Year'].astype(str)
index = 1
for by in data['Birth Year'] :
    data['Birth Year'][index] = dt.datetime.strptime(by,'%y').strftime('%Y')
    index += 1
data['Birth Year'] = data['Birth Year'].astype('int64')
```

```python
data.head()
```

```python
# Calculate 1st day value of each month
day_1= []
index = 1
for m in data['Birth Month'] :
    day_1.append(dt.date(data['Birth Year'][index],data['Birth Month'][index],1).weekday())
    index += 1
data['Day_1'] = day_1
data.head(2)
```

```python
data.info()
```

```python
data.Day = data.Day.astype('int64')
```

```python
# function to calculate birth date
# week_no - week in which birth day occurs
# day_no - Weekday representation from Monday=0 through Sunday=6
# day_1_no - Weekday representation for 1st of any given month
def calculate_birth_date(week_no,day_no,day_1_no) :
    birthDate = 7*week_no - (6-day_no)
    if(day_1_no > day_no) :
        birthDate += (7-day_1_no)
    else :
        birthDate -= day_1_no
    return birthDate
```

```python
# Calculate birth date
birth_date = []
index = 1
for d in data.Day :
    birth_date.append(dt.date(data['Birth Year'][index],data['Birth Month'][index],calculate_birth_date(data.Week[index],data.Day[index],data.Day_1[index])))
    index += 1
```

```python
data.insert(column = 'Birth Date', value = birth_date, loc =1)
data.head(2)
```

```python
#Changing birth date format
data['Birth Date'] = data['Birth Date'].astype('str')
index = 1
for by in data['Birth Date'] :
    data['Birth Date'][index] = dt.datetime.strptime(by,'%Y-%m-%d').strftime('%d-%m-%Y')
    index += 1
data.head(2)
```

```python
X = data.loc[:,['name','Birth Date','Gender','Height (in cms)','Bigger Foot','Dominant Hand']]
X.rename(columns ={'name' : 'Name'},inplace=True)
```
